```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
```

```python
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
train.head()
```

```python
combine = pd.concat((train.drop(['Survived'], axis=1), test), ignore_index=True)
combine.head()
```

```python
combine = combine.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
combine.head()
```

```python
combine.describe(include="all")
```

```python
sex_map = {'male':0, 'female':1}
combine['Sex'] = combine['Sex'].map(sex_map)
```

```python
combine.head()
```

```python
combine = combine.fillna({'Embarked':'S'})
combine['Embarked'].unique()
```

```python
cmb_map = {'S':1, 'C':2, 'Q':3}
combine['Embarked'] = combine['Embarked'].map(cmb_map)
combine.head()
```

```python
for x in range(len(combine["Fare"])):
    if pd.isnull(combine["Fare"][x]):
        pclass = combine["Pclass"][x] #Pclass = 3
        combine["Fare"][x] = round(combine[combine["Pclass"] == pclass]["Fare"].mean(), 4)
        
#map Fare values into groups of numerical values
combine['FareBand'] = pd.qcut(combine['Fare'], 4, labels = [1, 2, 3, 4])


#drop Fare values
combine = combine.drop(['Fare'], axis = 1)

combine.head()
```

```python
combine = combine.fillna({'Age':combine['Age'].median()})
```

```python
combine.head()
```

```python
combine['AgeBand'] = pd.qcut(combine['Age'], 6, labels = np.arange(6))
```

```python
combine = combine.drop(['Age'], axis=1)
combine.head()
```

```python
tarin_set = combine.head(train.shape[0])
test_set = combine[train.shape[0]:]
```

```python
from sklearn.model_selection import train_test_split
target = train['Survived']
x_train, x_val, y_train, y_val = train_test_split(tarin_set, target, test_size = 0.22, random_state = 0)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc = SVC()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_val)
acc_svc = round(accuracy_score(y_pred, y_val) * 100, 2)
print(acc_svc)
```

```python
# Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier

gbk = GradientBoostingClassifier()
gbk.fit(x_train, y_train)
y_pred = gbk.predict(x_val)
acc_gbk = round(accuracy_score(y_pred, y_val) * 100, 2)
print(acc_gbk)
```
